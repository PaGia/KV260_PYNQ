```python
import numpy as np
from pynq import allocate

# 1. Define the buffer clearly
# We use 'data_buffer' as the name here
data_buffer = allocate(shape=(1024,), dtype=np.int32)

# 2. Check the IP again to be safe
dma = ol.axi_dma_0

# 3. Try to receive data with a 'timeout' approach
# We don't use .wait() because it hangs if the hardware is dead
dma.recvchannel.transfer(data_buffer)

# Wait 1 second manually
import time
time.sleep(1)

# 4. Check results
print("Buffer Preview (First 10 samples):")
print(data_buffer[:10])

print("\nDMA Status Register:")
print(dma.register_map.SR_Status)
```

```python
# Force Reset the DMA Receive Channel
dma.recvchannel.stop()
dma.recvchannel.start()

print("DMA has been reset. Status should be Idle now.")
print(f"Is Idle: {dma.recvchannel.idle}")
```

```python
# Allocate a tiny buffer
tiny_buffer = allocate(shape=(10,), dtype=np.int32)

# Start transfer
dma.recvchannel.transfer(tiny_buffer)

# Wait a moment
import time
time.sleep(0.5)

if dma.recvchannel.idle:
    print("Success! Data received:", tiny_buffer)
else:
    print("Still no data. The pipeline is likely blocked upstream (DDS or SimpleNN).")
```
