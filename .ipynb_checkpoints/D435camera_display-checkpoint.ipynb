```python
import pyrealsense2 as rs
import cv2
import numpy as np
import os # 引入 os 模組用於路徑操作

# 定義儲存照片的路徑和檔案名稱
output_directory = "./realsense_photos"
color_image_filename = os.path.join(output_directory, "realsense_color_image.png")
depth_image_filename = os.path.join(output_directory, "realsense_depth_image.png")
combined_image_filename = os.path.join(output_directory, "realsense_combined_image.png")


try:
    # 建立輸出目錄，如果不存在的話
    os.makedirs(output_directory, exist_ok=True)
    print(f"確認輸出目錄: {output_directory}")

    # 配置深度和彩色串流
    pipeline = rs.pipeline()
    config = rs.config()
    # 設置解析度，這裡使用較常見的 640x480
    config.enable_stream(rs.stream.depth, 640, 480, rs.format.z16, 30)
    config.enable_stream(rs.stream.color, 640, 480, rs.format.bgr8, 30)

    # 啟動串流
    print("啟動 RealSense 串流...")
    pipeline.start(config)

    # 為了讓相機有時間自動曝光和調整，先跳過幾幀
    # 尤其是在第一次啟動時，通常需要幾幀來穩定
    for i in range(30): # 跳過約 1 秒的幀 (30 fps)
        pipeline.wait_for_frames()

    # 等待一組連貫的幀：深度和彩色
    print("捕捉一幀畫面...")
    frames = pipeline.wait_for_frames()
    depth_frame = frames.get_depth_frame()
    color_frame = frames.get_color_frame()

    if not depth_frame or not color_frame:
        print("錯誤：無法捕捉到有效的深度或彩色幀。")
    else:
        # 將影像轉換為 NumPy 陣列
        depth_image = np.asanyarray(depth_frame.get_data())
        color_image = np.asanyarray(color_frame.get_data())

        # (可選) 對深度影像應用色階映射，以便可視化
        depth_colormap = cv2.applyColorMap(cv2.convertScaleAbs(depth_image, alpha=0.03), cv2.COLORMAP_JET)

        # 將彩色影像和深度色階影像水平堆疊以便顯示和儲存
        combined_images = np.hstack((color_image, depth_colormap))

        # 儲存原始彩色影像
        cv2.imwrite(color_image_filename, color_image)
        print(f"彩色影像已儲存至：{color_image_filename}")

        # 儲存深度色階影像
        cv2.imwrite(depth_image_filename, depth_colormap)
        print(f"深度影像（色階）已儲存至：{depth_image_filename}")

        # 儲存結合後的影像
        cv2.imwrite(combined_image_filename, combined_images)
        print(f"結合影像（彩色+深度）已儲存至：{combined_image_filename}")

        # (可選) 顯示捕捉到的影像 (如果你正在使用 X11 轉發或有顯示器)
        print("在視窗中顯示捕捉到的影像 (按任意鍵關閉)...")
        cv2.imshow('Captured RealSense Image (Color | Depth)', combined_images)
        cv2.waitKey(0) # 等待按鍵，直到使用者關閉視窗

finally:
    # 停止串流
    print("停止 RealSense 串流。")
    pipeline.stop()
    cv2.destroyAllWindows()
    print("程式執行結束。")
```
